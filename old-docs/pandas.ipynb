## Cambiar el orden de las columnas de un marco de datos de Pandas

## ¿Cómo ordenar un marco de datos de pandas?

Puede usar la función pandas dataframe sort_values() para ordenar un marco de datos. Permite la flexibilidad de clasificar un marco de datos por una o más columnas, elegir el algoritmo de clasificación, cómo tratar los NaN durante las comparaciones, usar una clave personalizada para clasificar, etc. La siguiente es la sintaxis:

df.sort_values (por, ascending = True, in_place = True)

Pase la columna o lista de columnas para ordenar al parámetro por. De forma predeterminada, devuelve un marco de datos ordenado y no altera el marco de datos original. Si desea modificar el marco de datos original in situ, pase inplace=True

Seguiremos con el dataframe previamente definido

## 1. Ordenar marco de datos por una sola columna

Para ordenar un marco de datos por una sola columna, pase el nombre de la columna al parámetro by de la función sort_values(). Por ejemplo, para ordenar el marco de datos anterior por Altura:

```python
df_sorted = df.sort_values(by='Shares',ascending=False)
print(df_sorted)
```

```python
df
```

## 2. Ordenar marco de datos por múltiples columnas

También puede ordenar un marco de datos de pandas por varias columnas. Para esto, pase las columnas por las que desea ordenar el marco de datos como una lista al parámetro por. Ejemplo, para ordenar el marco de datos df por altura y campeonatos:

```python
df_sorted = df.sort_values(by=['Shares','Symbol'],ascending=[False,True])
print(df_sorted)
```

## 3. Ordene el marco de datos poniendo NaNs primero

Si desea ordenar un marco de datos por una columna que tiene uno o más valores de NaN, puede decidir ordenarlos primero o último. Este comportamiento está controlado por el parámetro na_position que toma los valores 'first' y 'last' y es 'last' por defecto. Aquí hay un ejemplo:

Primero, creamos una copia del marco de datos original y establecemos los shares de Apple en NaN.

```python
import numpy as np
df_na = df.copy()
df_na.loc[3, 'Shares'] = np.nan
print(df_na)
```

Luego, ordenamos el marco de datos df_na por la altura de la columna de modo que los NaN se coloquen primero en la clasificación.

```python
df_na[df_na.isna()]=np.mean(df_na.Shares)
```

```python
df_na
```

## 4. Ordene el marco de datos por un algoritmo de clasificación diferente
La función sort_values() también le permite elegir entre tres algoritmos de clasificación diferentes. El tipo de parámetro controla este comportamiento. Toma 'quicksort', 'mergesort' y 'heapsort' como valores y es 'quicksort' por defecto. Por ejemplo, para ordenar un marco de datos por una columna usando 'mergesort' como algoritmo de clasificación:

```python
df_sorted = df.sort_values(by='Shares', kind='mergesort')
print(df_sorted)
```

## Pandas DataFrame a una lista en Python
Los marcos de datos de Pandas son excelentes para manipular datos. Pero, a veces, puede suceder que prefiera tener los datos como una lista (o más precisamente, una lista de listas). En este tutorial, veremos cómo convertir un marco de datos de pandas en una lista de python.

## ¿Cómo convertir un marco de datos de pandas en una lista?
Hay varias formas de obtener una lista de python a partir de un marco de datos de pandas, según el tipo de lista que desee crear. Para obtener rápidamente una lista de un marco de datos con cada elemento que representa una fila en el marco de datos, puede usar la función tolist() como df.values.tolist()

Sin embargo, también hay otras formas. Puede crear una lista con cada elemento que represente una columna de marco de datos. O bien, puede crear algo muy específico en función de sus requisitos. Veamos algunos de los diferentes casos de uso con ejemplos.

## Ejemplos
Siguiendo con nuestro dataframe

### 1. Lista con filas de DataFrame como elementos

Como se mencionó anteriormente, puede obtener rápidamente una lista de un marco de datos utilizando la función tolist().

```python
ls = df.values.tolist()
print(ls)
```

### 2. Lista con columnas DataFrame como elementos
También puede usar la función tolist() en columnas individuales de un marco de datos para obtener una lista con valores de columna.

```python
# list with each item representing a column
ls = []
for col in df.columns:
    # convert pandas series to list
    col_ls = df[col].tolist()
    # append column list to ls
    ls.append(col_ls)
# print the created list
print(ls)
```

En el ejemplo anterior, iteramos a través de cada columna del marco de datos que se convierte en una lista y luego se agrega a ls. Puede ver que aquí obtenemos una lista de listas con cada elemento de la lista que representa una columna en el marco de datos.

### 3. Lista de un DataFrame iterando a través de las filas
También puede crear una lista iterando a través de las filas del marco de datos.

```python
ls = []
# iterate over the rows
for i, row in df.iterrows():
    # create a list representing the dataframe row
    row_ls = [row['Name'], row['Industry'],row['Shares'], row['Symbol']]
    # append row list to ls
    ls.append(row_ls)

print(ls)
```

En el ejemplo anterior, usamos la función pandas dataframe iterrows() para iterar sobre las filas de df y crear una lista con valores de fila que se agrega a ls. Puede ver que obtenemos una lista de listas con cada elemento de la lista que representa una fila en el marco de datos como vimos en el ejemplo con la función tolist().

Este método también le permite la flexibilidad de crear listas específicas según sus requisitos. Por ejemplo, a partir del marco de datos anterior, si desea crear una lista de listas con solo el símbolo de la acción y su respectivo recuento de acciones, puede hacerlo fácilmente conservando solo esos campos.

```python
ls = []
# iterate over the rows
for i, row in df.iterrows():
    # create a list representing the dataframe row
    row_ls = [row['Symbol'], row['Shares']]
    # append row list to ls
    ls.append(row_ls)

print(ls)
```

## Recuento de valores únicos en cada columna

Generalmente, los datos en cada columna representan una característica diferente de un marco de datos de pandas. Puede ser continuo, categórico o algo totalmente diferente como textos distintos. Si no está seguro acerca de la naturaleza de los valores con los que está tratando, podría ser un buen paso exploratorio conocer el recuento de valores distintos. En este tutorial, veremos cómo obtener el recuento de valores únicos en cada columna de un marco de datos de pandas.

La función *núnica()*
Para contar los valores únicos de cada columna de un marco de datos, puede usar la función pandas dataframe nunique(). La siguiente es la sintaxis:

*cuenta = df.nunique()*

Aquí, #df# es el marco de datos para el que desea conocer los recuentos únicos. Devuelve una serie de pandas de conteos. De forma predeterminada, la función pandas dataframe nunique() cuenta los valores distintos a lo largo del eje = 0, es decir, en filas, lo que le brinda el recuento de valores distintos en cada columna.

Ejemplos

Veamos algunos de los diferentes casos de uso para obtener recuentos únicos a través de algunos ejemplos. Primero, crearemos un marco de datos de muestra que usaremos a lo largo de este tutorial.

```python
import pandas as pd
import numpy as np
# create a sample dataframe
data = {
    'EmpCode': ['E1', 'E2', 'E3', 'E4', 'E5'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'Age': [27, 24, 29, 24, 25],
    'Department': ['Accounting', 'Sales', 'Accounting', np.nan, 'Sales']
}
df = pd.DataFrame(data)
# display the dataframe
df
```

El uso de la función pandas dataframe nunique() con parámetros predeterminados proporciona un recuento de todos los valores distintos en cada columna.

## Eliminar duplicados de un marco de datos de Pandas

Mientras trabaja con datos, puede haber situaciones en las que su marco de datos tenga filas duplicadas. Saber cómo eliminar dichas filas rápidamente puede ser muy útil. En este tutorial, veremos cómo eliminar duplicados de un marco de datos de pandas a través de algunos ejemplos.

### La función drop_duplicates()
La función pandas dataframe drop_duplicates() se puede usar para eliminar filas duplicadas de un marco de datos. También le brinda la flexibilidad de identificar duplicados en función de ciertas columnas a través del parámetro de subconjunto. La siguiente es su sintaxis:

df.drop_duplicates()

Devuelve un marco de datos con las filas duplicadas eliminadas. Elimina los duplicados excepto la primera aparición de forma predeterminada. Puede cambiar este comportamiento a través del parámetro keep que incluye 'primero', 'último' o Falso. Para modificar el marco de datos en el lugar, pase el argumento inplace=True.

### Ejemplos

Veamos algunos de los casos de uso de la función drop_duplicates() a través de ejemplos:

### 1. Eliminar filas duplicadas en función de todas las columnas
De forma predeterminada, la función drop_duplicates() identifica los duplicados teniendo en cuenta todas las columnas. Luego, elimina las filas duplicadas y solo mantiene su primera aparición.


```python
import pandas as pd

# create a sample dataframe with duplicate rows
data = {
    'Pet': ['Cat', 'Dog', 'Dog', 'Dog', 'Cat'],
    'Color': ['Brown', 'Golden', 'Golden', 'Golden', 'Black'],
    'Eyes': ['Black', 'Black', 'Black', 'Brown', 'Green']
}

df = pd.DataFrame(data)

# print the dataframe
print("The original dataframe:\n")
print(df)

# drop duplicates
df_unique = df.drop_duplicates()
print("\nAfter dropping duplicates:\n")
print(df_unique)
```

En el ejemplo anterior, puede ver que las filas con índice 1 y 2 tienen los mismos valores para las tres columnas. Al aplicar la función drop_duplicates(), se conserva la primera fila y se descartan las restantes filas duplicadas. Como resultado, el marco de datos devuelto no tiene un índice continuo. Si desea que el marco de datos devuelto tenga un índice continuo, pase ignore_index=True a la función drop_duplicates() o restablezca el índice del marco de datos devuelto.

### 2. Suelte filas duplicadas en función de ciertas columnas

También puede indicar a la función drop_duplicates() que identifique los duplicados basándose solo en ciertas columnas pasándolas como una lista al argumento del subconjunto.

```python
import pandas as pd

# create a sample dataframe with duplicate rows
data = {
    'Pet': ['Cat', 'Dog', 'Dog', 'Dog', 'Cat'],
    'Color': ['Brown', 'Golden', 'Golden', 'Golden', 'Black'],
    'Eyes': ['Black', 'Black', 'Black', 'Brown', 'Green']
}

df = pd.DataFrame(data)

# print the dataframe
print("The original dataframe:\n")
print(df)

# drop duplicates
df_unique = df.drop_duplicates(subset=['Pet', 'Color'])
print("\nAfter dropping duplicates:\n")
print(df_unique)
```

En el ejemplo anterior, identificamos los duplicados basándonos solo en las columnas Pet y Color pasándolos como una lista a la función drop_duplicates(). Con este criterio, las filas con índice 1, 2 y 3 ahora son duplicados y el marco de datos devuelto solo conserva la primera fila.

### 3. Elimina los duplicados y conserva la última aparición

Si desea conservar la última fila duplicada en lugar de la primera, pase keep='last' a la función drop_duplicates().

```python
import pandas as pd

# create a sample dataframe with duplicate rows
data = {
    'Pet': ['Cat', 'Dog', 'Dog', 'Dog', 'Cat'],
    'Color': ['Brown', 'Golden', 'Golden', 'Golden', 'Black'],
    'Eyes': ['Black', 'Black', 'Black', 'Brown', 'Green']
}

df = pd.DataFrame(data)

# print the dataframe
print("The original dataframe:\n")
print(df)

# drop duplicates
df_unique = df.drop_duplicates(keep='last')
print("\nAfter dropping duplicates:\n")
print(df_unique)
```

En el ejemplo anterior, conservamos el último duplicado en lugar del primero.

### 4. Elimine los duplicados y no conserve ninguna ocurrencia
Si no desea conservar ninguna de las filas duplicadas, pase keep=False a la función drop_duplicates().

```python
import pandas as pd

# create a sample dataframe with duplicate rows
data = {
    'Pet': ['Cat', 'Dog', 'Dog', 'Dog', 'Cat'],
    'Color': ['Brown', 'Golden', 'Golden', 'Golden', 'Black'],
    'Eyes': ['Black', 'Black', 'Black', 'Brown', 'Green']
}

df = pd.DataFrame(data)

# print the dataframe
print("The original dataframe:\n")
print(df)

# drop duplicates
df_unique = df.drop_duplicates(keep=False)
print("\nAfter dropping duplicates:\n")
print(df_unique)
```

## Concatenar Marcos de datos en Pandas

Puede concatenar marcos de datos a lo largo de la fila o el eje de la columna. Digamos que tiene varios marcos de datos con los mismos campos y le gustaría combinarlos en uno. En este caso, querrá concatenar los marcos de datos a lo largo del índice o el eje de la fila. O digamos, si tiene algunos campos adicionales para sus datos actuales que desee agregar, entonces querrá concatenarlos a lo largo del eje de la columna. En este tutorial, veremos cómo concatenar dos o más marcos de datos en pandas.

La función pandas concat()
La función pandas concat() se usa para concatenar múltiples marcos de datos en uno. La siguiente es su sintaxis:

pd.concat(objetos, eje=0)
Pasa la secuencia de objetos de marcos de datos (objs) que desea concatenar e indica el eje (0 para filas y 1 para columnas) a lo largo del cual se realizará la concatenación y devuelve el marco de datos concatenado. El valor predeterminado para el eje es 0. Tenga en cuenta que también puede usar la función concat para concatenar series de pandas.

Ejemplos
Veamos algunos de los casos de uso de la función concat() a través de ejemplos:

### 1. Concatenar DataFrames con los mismos campos (verticalmente)
Si tiene dos o más marcos de datos que tienen los mismos campos (o columnas) y desea concatenarlos en uno:


```python
import pandas as pd

# Create two dataframes with same columns
df1 = pd.DataFrame({'Name': ['Sam', 'Emma'], 'Age': [14, 15]})
df2 = pd.DataFrame({'Name': ['Karen', 'Rahul'], 'Age': [10, 13]})

# Print the dataframes
print("DataFrame df1:\n")
print(df1)
print("\nDataFrame df2:\n")
print(df2)

# Concatenate them along the axis
df_combined = pd.concat([df1, df2])

# Print the concatenated dataframe
print("\nCombined DataFrame:\n", df_combined)
```

In the above example, the dataframes df1 and df2 have the same columns – Name and Age and are concatenated along the index, that is vertically (since the axis parameter is by default 0). Also, note that the combined dataframe retains the index from the individual dataframes, you can reset the index by using the reset_index() function.

```python
df_combined = df_combined.reset_index(drop=True)
print(df_combined)
```

Lo anterior es el marco de datos combinado con su restablecimiento de índice.

### 2. Concatenar DataFrames con filas que se refieren a los mismos registros (horizontalmente)

Si tiene dos o más marcos de datos con filas que se refieren a los mismos registros y desea concatenarlos para tener todos los campos diferentes en un solo marco de datos, use la función concat() con axis=1.

```python
import pandas as pd

# Create two dataframes with same columns
df1 = pd.DataFrame({'Name': ['Sam', 'Emma'], 'Age': [14, 15]})
df2 = pd.DataFrame({'Math': ['B', 'A+'], 'Science': ['A', 'B+']})

# Print the dataframes
print("DataFrame df1:\n")
print(df1)
print("\nDataFrame df2:\n")
print(df2)

# Concatenate them along the axis
df_combined = pd.concat([df1, df2], axis=1)

# Print the concatenated dataframe
print("\nCombined DataFrame:\n", df_combined)
```

En el ejemplo anterior, los marcos de datos df1 (que almacenan el nombre y la edad de los estudiantes) y df2 (que almacenan sus calificaciones en Matemáticas y Ciencias) se concatenan horizontalmente (a lo largo del eje de la columna).

## Soltar una o más columnas de un marco de datos

Los marcos de datos de Pandas son bastante poderosos para manipular datos. A menudo, mientras trabaja con datos, particularmente durante EDA (Análisis exploratorio de datos) y el preprocesamiento de datos, es posible que deba eliminar una o más columnas. En este tutorial, veremos cómo eliminar una o más columnas de un marco de datos de pandas con algunos ejemplos.

### ¿Cómo eliminar columnas de un marco de datos de pandas?

Puede usar la función pandas dataframe drop() con el eje establecido en 1 para eliminar una o más columnas de un marco de datos. La siguiente es la sintaxis:

df.drop(cols_to_drop, eje=1)

Aquí, cols_to_drop es el índice o las etiquetas de las columnas que se van a soltar; si se van a soltar más de una columna, debería ser una lista. El eje representa el eje del que eliminar las etiquetas; el valor predeterminado es 0, pero si desea eliminar columnas, pase el eje como 1 (es decir, 0 para filas y 1 para columnas).

También tenga en cuenta que la función drop() no modifica el marco de datos en el lugar de forma predeterminada. Devuelve una copia del marco de datos con las etiquetas eliminadas. Si desea modificar el marco de datos en el lugar, pase el argumento inplace=True a la función.

### Soltar columnas por nombre
Para eliminar columnas por nombre, simplemente pase el nombre de la columna (si desea eliminar una sola columna) o la lista de columnas (si desea eliminar varias columnas) a la función de eliminación. Vea los ejemplos a continuación:

### Ejemplo 1: suelte una sola columna por nombre

```python
import pandas as pd

# create a sample dataframe
data = {
    'A': ['a1', 'a2', 'a3'],
    'B': ['b1', 'b2', 'b3'],
    'C': ['c1', 'c2', 'c3'],
    'D': ['d1', 'd2', 'd3']
}

df = pd.DataFrame(data)

# print the dataframe
print("Original Dataframe:\n")
print(df)

# remove column C
df = df.drop('C', axis=1)

print("\nAfter dropping C:\n")
print(df)
```

En el ejemplo anterior, se crea un marco de datos de muestra df con cuatro columnas A, B, C y D. Luego, la columna C se elimina mediante la función drop(). Tenga en cuenta que, dado que tuvimos que eliminar solo una columna, no necesitábamos pasar una lista.

### Ejemplo 2: Deshechar varias columnas por nombre

```python
import pandas as pd

# create a sample dataframe
data = {
    'A': ['a1', 'a2', 'a3'],
    'B': ['b1', 'b2', 'b3'],
    'C': ['c1', 'c2', 'c3'],
    'D': ['d1', 'd2', 'd3']
}

df = pd.DataFrame(data)

# print the dataframe
print("Original Dataframe:\n")
print(df)

# remove columns C and D
df = df.drop(['C', 'D'], axis=1)

print("\nAfter dropping columns C and D:\n")
print(df)
```

En el ejemplo anterior, las columnas C y D se eliminan del marco de datos df. Tenga en cuenta que tuvimos que proporcionar la lista de nombres de columnas para descartar ya que estábamos colocando varias columnas juntas.

### Deshechar columnas por índice

Para eliminar columnas por número de columna, pase df.columns[i] a la función drop() donde i es el índice de columna de la columna que desea eliminar. Para eliminar varias columnas por sus índices, pase df.columns[[i, j, k]] donde i, j, k son los índices de columna de las columnas que desea eliminar.

### Ejemplo 1: deshechar una sola columna por índice

```python
import pandas as pd

# create a sample dataframe
data = {
    'A': ['a1', 'a2', 'a3'],
    'B': ['b1', 'b2', 'b3'],
    'C': ['c1', 'c2', 'c3'],
    'D': ['d1', 'd2', 'd3']
}

df = pd.DataFrame(data)

# print the dataframe
print("Original Dataframe:\n")
print(df)

# remove column
df = df.drop(df.columns[2], axis=1)

print("\nAfter dropping C:\n")
print(df)
```

En el ejemplo anterior, la columna C se elimina utilizando su índice 2 del marco de datos df.

Ejemplo 2: suelte varias columnas con su índice

```python
import pandas as pd

# create a sample dataframe
data = {
    'A': ['a1', 'a2', 'a3'],
    'B': ['b1', 'b2', 'b3'],
    'C': ['c1', 'c2', 'c3'],
    'D': ['d1', 'd2', 'd3']
}

df = pd.DataFrame(data)

# print the dataframe
print("Original Dataframe:\n")
print(df)

# remove columns C and D
df = df.drop(df.columns[[2, 3]], axis=1)

print("\nAfter dropping columns C and D:\n")
print(df)
```

En el ejemplo anterior, las columnas con índice 2 y 3 se eliminan del marco de datos df.

Detrás de escena, df.columns en realidad da el nombre de la columna para el índice dado:
