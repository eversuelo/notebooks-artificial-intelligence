# Simple Linear Regression

Simple linear regression is one of the most fundamental and widely used models in Machine Learning and statistics. It allows us to understand and model the relationship between a dependent variable and one or more independent variables. This relationship is represented by a straight line that tries to fit the observed data as closely as possible by minimizing the error between predictions and real values.

The linear regression model assumes that there is a linear relationship between the variables, meaning that the change in the dependent variable is proportional to the change in the independent variable. In its most basic form, simple linear regression is expressed as:

$$
y=b_0+b_1*x_1
$$

where:
- **y** is the dependent variable we want to predict.
- **x** is the independent variable used to make the prediction.
- **$b_0$** is the intercept, or the value of y when x is equal to zero.
- **$b_1$** is the slope of the line, which indicates the rate of change of y with respect to x.
- **ε** is the error term, representing the difference between the predicted value and the true value due to factors not included in the model.

In Machine Learning, linear regression is used to make numerical predictions and also as an initial tool for exploring relationships between variables. It is an interpretable and easy-to-implement model, which makes it ideal for beginners. Through algorithms such as the least squares method, the line is adjusted to minimize the sum of squared errors, providing an optimal solution for the available dataset.

However, linear regression has some limitations. The main one is the linearity assumption, which does not always hold in more complex situations. It is also sensitive to outliers, since a few extreme observations can distort the fitted line. For this reason, more advanced models are often used in complex problems.

Despite these limitations, linear regression remains a fundamental tool for understanding prediction and building intuition about data behavior. Its simplicity and interpretability make it a natural starting point for Machine Learning.

## Types of Variables
In simple linear regression, variables can mainly be classified into two types: numeric variables and categorical variables.

### Numeric Variables
Numeric variables represent measurable quantities and can take continuous or discrete values. Common examples include monthly income, temperature, age, and price. In the context of linear regression, these variables are used directly to model the relationship between a dependent variable and an independent variable.

### Categorical Variables
Categorical variables do not represent numeric quantities but instead indicate groups or categories. For example, a car's color, the type of city, or a person's gender are categorical variables. These variables must be transformed before they can be used in a linear regression model, because regression algorithms require numeric values. A common method is dummy-variable encoding, in which each category is represented by a binary value.

## Regression and Its Types
Regression is a statistical technique used to model the relationship between a dependent variable and one or more independent variables. Different regression models are useful for different problems.

### Simple Linear Regression
Used when we want to model the relationship between a dependent variable and a single independent variable.

### Multiple Linear Regression
An extension of simple linear regression that allows more than one independent variable.

### Simple Logistic Regression
Used when the dependent variable is binary, such as yes/no or positive/negative.

### Multiple Logistic Regression
Used when the dependent variable is binary and there is more than one independent variable.

## Least Squares Method
The least squares method is a mathematical technique used to find the best possible approximation to a series of observed data by fitting a function, usually a straight line, that minimizes the sum of squared errors.

In simple linear regression, the relationship between a dependent variable ($y$) and an independent variable ($x$) is modeled by a straight line of the form:

\[ y = b_0 + b_1x \]

where:
- **$b_0$** is the intercept.
- **$b_1$** is the slope of the line.

The least squares method seeks the values of $b_0$ and $b_1$ that minimize the sum of the squared differences between the observed values ($y_i$) and the predicted values ($\hat{y}_i$):

$$
S = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - (b_0 + b_1 x_i))^2
$$

This provides a strong foundation for understanding relationships between variables and making predictions with clear interpretation.


# Practical Case
### Practical Case: Predicting Salaries from Years of Experience

Suppose you work in the human resources department of a company and want to understand the relationship between employees' work experience and their salaries. You have a dataset containing information about years of experience and salary. Your goal is to predict the salary of new employees based on their experience using simple linear regression.

In this practical case:
- **Independent Variable (X)**: Years of work experience.
- **Dependent Variable (Y)**: Employee salary.

The problem consists of building a simple linear regression model to determine the relationship between years of experience and salary. The model finds the best straight line that represents the trend of the data and can be used to make predictions for new employees.

### Applying the Model
With the available data on "Years of Experience" and "Salary", we can train a linear regression model that finds the equation of a straight line that best fits the data:

\[ y = b_0 + b_1 x \]

where:
- $y$ is the predicted salary.
- $x$ is the number of years of experience.
- $b_0$ is the salary value when years of experience are zero.
- $b_1$ is the slope of the line.

### Use in the Company
This model can be useful for:
- **Hiring**: estimating a competitive salary offer based on prior experience.
- **Evaluations and promotions**: supporting more consistent salary decisions.
- **Budgeting**: estimating payroll needs for future hiring.

### Conclusion
Simple linear regression is a valuable tool for analyzing and predicting the effect of one variable on another. In this case, it models the relationship between years of experience and salary and gives us a clear estimate of how salary changes with experience.


```python
# Simple Linear Regression

# 1) Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 2) Import the dataset (use the real path + column names, not positional indices)
dataset = pd.read_csv('data/salary_data.csv')
print("First rows:\n", dataset.head(), "\n")
print("Shape:", dataset.shape)

X = dataset[['YearsExperience']].values   # feature matrix (2D, as scikit-learn expects)
y = dataset['Salary'].values              # target vector (1D)

# 3) Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1/3, random_state=0
)

# NOTE: feature scaling is NOT needed for ordinary least-squares LinearRegression,
# so we deliberately skip StandardScaler here.

# 4) Train the simple linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# 5) Inspect the learned line  ->  y = b0 + b1 * x
b0 = regressor.intercept_
b1 = regressor.coef_[0]
print(f"\nFitted line:  Salary = {b0:,.2f} + {b1:,.2f} * YearsExperience")

# 6) Predict and evaluate on the (unseen) test set
y_pred = regressor.predict(X_test)
print("\n=== Test-set performance ===")
print(f"R^2  : {r2_score(y_test, y_pred):.4f}")
print(f"MAE  : {mean_absolute_error(y_test, y_pred):,.2f}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, y_pred)):,.2f}")

# Side-by-side comparison of actual vs predicted salaries
comparison = pd.DataFrame({
    'YearsExperience': X_test.ravel(),
    'Actual': y_test,
    'Predicted': y_pred.round(2),
    'Error': (y_test - y_pred).round(2),
}).sort_values('YearsExperience').reset_index(drop=True)
print("\nActual vs Predicted (test set):")
print(comparison)

# 7) Predict the salary for a brand-new employee
new_experience = np.array([[5.0]])
predicted_salary = regressor.predict(new_experience)[0]
print(f"\nPredicted salary for 5 years of experience: ${predicted_salary:,.2f}")

# 8) Visualize training and test results side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(X_train, y_train, color="red", label="Actual")
axes[0].plot(X_train, regressor.predict(X_train), color="blue", label="Regression line")
axes[0].set_title("Salary vs Years of Experience (Training Set)")
axes[0].set_xlabel("Years of Experience")
axes[0].set_ylabel("Salary (in $)")
axes[0].legend()

axes[1].scatter(X_test, y_test, color="green", label="Actual")
axes[1].plot(X_train, regressor.predict(X_train), color="blue", label="Regression line")
axes[1].set_title("Salary vs Years of Experience (Test Set)")
axes[1].set_xlabel("Years of Experience")
axes[1].set_ylabel("Salary (in $)")
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Interpreting the Results

Once the model is trained, the code above reports several quantities that tell us *how good* the fit is and *what it means*:

### The fitted line — $y = b_0 + b_1 x$
- **`intercept_` ($b_0$)** is the predicted salary for **zero** years of experience.
- **`coef_` ($b_1$)** is the **slope**: how much the salary increases, on average, for **each additional year** of experience.

### Evaluation metrics (computed on the unseen *test* set)
| Metric | Meaning | Better when |
|---|---|---|
| **R²** (coefficient of determination) | Fraction of the variance in salary explained by experience (0 = none, 1 = perfect) | **closer to 1** |
| **MAE** (Mean Absolute Error) | Average absolute difference between predicted and real salary, in dollars | **lower** |
| **RMSE** (Root Mean Squared Error) | Like MAE but penalizes large errors more; same units as salary | **lower** |

> We evaluate on the **test set** — data the model never saw during training — because measuring error on the training data would be overly optimistic and hide overfitting.

### The Actual vs Predicted table
The `Error` column (`Actual − Predicted`) shows where the model over- or under-estimates. Small, balanced errors around zero indicate a healthy fit.
